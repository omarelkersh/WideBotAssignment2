# Encoding author column

```python
num_unique_values = merged_df['author'].nunique()
unique_values = merged_df['author'].unique()
print(unique_values)
print(num_unique_values)
merged_df['author_Encoded'] = pd.factorize(merged_df['author'])[0]
```

```python
merged_df.corr()
```

# P-value

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(merged_df['author_Encoded'], merged_df['topic_Encoded'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

since p-value is less than <0.001 then there is a strong correlation between topic and auhtor
